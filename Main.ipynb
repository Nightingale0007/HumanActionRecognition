# VGG

## HMBD51

```python
import vgg.data_process as data_process
```

```python
ori_data_path = 'HMBD51_videos'
out_label_path = 'vgg/data'
output_data_path = 'vgg/data/HMBD51_image'

# Generate label documentation
data_process.label_text_write(ori_data_path, out_label_path)
# Partition the dataset to generate the corresponding image dataset.
data_process.preprocess(ori_data_path, output_data_path)
```

```python
from vgg.train import train
```

```python
train_path = "vgg/data/HMBD51_image/train"
val_path = "vgg/data/HMBD51_image/val" 
test_path = "vgg/data/HMBD51_image/test"

epochs = 30
learning_rate = 1e-3
save_dir = 'vgg/vgg_hmbd51_model.pth'


trained_model, final_val_accuracy, test_accuracy, class_accuracies = train(
    train_data_path=train_path,
    val_data_path=val_path,
    test_data_path=test_path,
    epochs=epochs,
    learning_rate=learning_rate,
    model_save_path=save_dir
)

print(f"\n" + "="*50)
print("TRAINING SUMMARY")
print("="*50)
print(f"Final Validation Accuracy: {final_val_accuracy:.4f}")
print(f"Test Set Accuracy: {test_accuracy:.4f}")
print(f"Average Class Accuracy: {sum(class_accuracies.values()) / len(class_accuracies):.4f}")
```

```python
import torch
from torch.utils.data import DataLoader
from c3d.dataset import VideoDataset
from c3d.train import train_model
```

```python
device = torch.device("cuda:0" if torch.cuda.is_available() else "cpu")
num_epochs = 30  # Training session
num_classes = 3   # The dataset utilised by the model and the output parameters of the final layer of the neural network
lr = 1e-3  # Learning rate
save_dir = 'c3d/c3d_hmbd51_model.pth'

train_dataloader = DataLoader(VideoDataset(dataset_path='c3d\data\HMBD51_image', images_path='train', clip_len=16), batch_size=16, shuffle=True, num_workers=2)
val_dataloader = DataLoader(VideoDataset(dataset_path='c3d\data\HMBD51_image', images_path='val', clip_len=16), batch_size=16, num_workers=2)
test_dataloader = DataLoader(VideoDataset(dataset_path='c3d\data\HMBD51_image', images_path='test', clip_len=16), batch_size=16, num_workers=2)

train_model(num_epochs, num_classes, lr, device, save_dir, train_dataloader, val_dataloader, test_dataloader)
```
